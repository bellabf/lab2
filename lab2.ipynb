```python
import json
import pandas as pd
import numpy as np
from datetime import datetime
from pathlib import Path

# ============================================================
# CONFIG
# ============================================================
JSON_FILE = Path('data/titanic_data.json')
JSON_FILE.parent.mkdir(parents=True, exist_ok=True)  # ✅ Creates 'data/' folder if missing


# ============================================================
# CLASS 1: Passenger
# ============================================================
class Passenger:
    """Represents a single passenger with all attributes."""

    def __init__(self, passenger_id, name, age, sex, survived, pclass,
                 fare, embarked=None, sibsp=None, parch=None,
                 family_size=None, is_alone=None, age_group=None, title=None):

        self.passenger_id = int(passenger_id)       if pd.notna(passenger_id) else None
        self.name         = str(name)               if pd.notna(name)         else None
        self.age          = round(float(age), 1)    if pd.notna(age)          else None
        self.sex          = str(sex)                if pd.notna(sex)          else None
        self.survived     = int(survived)           if pd.notna(survived)     else None
        self.pclass       = int(pclass)             if pd.notna(pclass)       else None
        self.fare         = round(float(fare), 2)   if pd.notna(fare)         else None
        self.embarked     = str(embarked)           if pd.notna(embarked)     else None
        self.sibsp        = int(sibsp)              if pd.notna(sibsp)        else None
        self.parch        = int(parch)              if pd.notna(parch)        else None
        self.family_size  = int(family_size)        if pd.notna(family_size)  else None
        self.is_alone     = int(is_alone)           if pd.notna(is_alone)     else None
        self.age_group    = str(age_group)          if pd.notna(age_group)    else None
        self.title        = str(title)              if pd.notna(title)        else None

    def to_dict(self):
        """Convert passenger to dictionary for JSON export."""
        return {
            'passenger_id': self.passenger_id,
            'name':         self.name,
            'title':        self.title,
            'age':          self.age,
            'age_group':    self.age_group,
            'sex':          self.sex,
            'survived':     self.survived,
            'pclass':       self.pclass,
            'fare':         self.fare,
            'embarked':     self.embarked,
            'sibsp':        self.sibsp,
            'parch':        self.parch,
            'family_size':  self.family_size,
            'is_alone':     self.is_alone,
        }


# ============================================================
# CLASS 2: TitanicDataset
# ============================================================
class TitanicDataset:
    """Encapsulates all data processing: cleaning, stats, feature engineering, export."""

    def __init__(self, dataframe):
        self.raw_df     = dataframe.copy()
        self.df         = dataframe.copy()
        self.passengers = []

        # Run all processing steps on init
        self._engineer_features()
        self._create_passengers()

    # ----------------------------------------------------------
    # STEP 1: Feature Engineering (from previous work)
    # ----------------------------------------------------------
    def _categorize_age(self, age):
        if pd.isna(age):    return 'Unknown'
        elif age < 18:      return 'Child'
        elif age < 30:      return 'Young Adult'
        elif age < 50:      return 'Adult'
        else:               return 'Senior'

    def _engineer_features(self):
        """Create FamilySize, IsAlone, AgeGroup features."""
        self.df['FamilySize'] = self.df['SibSp'] + self.df['Parch'] + 1
        self.df['IsAlone']    = (self.df['FamilySize'] == 1).astype(int)
        self.df['AgeGroup']   = self.df['Age'].apply(self._categorize_age)
        print("✅ Feature engineering complete (FamilySize, IsAlone, AgeGroup)")

    # ----------------------------------------------------------
    # STEP 2: Build Passenger objects
    # ----------------------------------------------------------
    def _create_passengers(self):
        """Instantiate a Passenger object for every row."""
        for idx, row in self.df.iterrows():
            p = Passenger(
                passenger_id = row.get('PassengerId', idx),
                name         = row.get('Name',        None),
                age          = row.get('Age',         None),
                sex          = row.get('Sex',         None),
                survived     = row.get('Survived',    None),
                pclass       = row.get('Pclass',      None),
                fare         = row.get('Fare',        None),
                embarked     = row.get('Embarked',    None),
                sibsp        = row.get('SibSp',       None),
                parch        = row.get('Parch',       None),
                family_size  = row.get('FamilySize',  None),
                is_alone     = row.get('IsAlone',     None),
                age_group    = row.get('AgeGroup',    None),
                title        = row.get('Title',       None),
            )
            self.passengers.append(p)
        print(f"✅ Created {len(self.passengers)} Passenger objects")

    # ----------------------------------------------------------
    # STEP 3: Missing Values Analysis (from previous work)
    # ----------------------------------------------------------
    def get_missing_values(self):
        """Return missing value counts and percentages per column."""
        missing_count   = self.raw_df.isnull().sum()
        missing_percent = (missing_count / len(self.raw_df) * 100).round(2)
        missing_df = pd.DataFrame({
            'missing_count':   missing_count,
            'missing_percent': missing_percent
        })
        return missing_df[missing_df['missing_count'] > 0].sort_values('missing_count', ascending=False).to_dict()

    # ----------------------------------------------------------
    # STEP 4: Statistics (from previous work)
    # ----------------------------------------------------------
    def get_statistics(self):
        """Return mean, median, std for all numeric columns."""
        numeric_df = self.df.select_dtypes(include='number')
        return {
            'mean':   numeric_df.mean().round(2).to_dict(),
            'median': numeric_df.median().round(2).to_dict(),
            'std':    numeric_df.std().round(2).to_dict(),
        }

    # ----------------------------------------------------------
    # STEP 5: Summary Statistics
    # ----------------------------------------------------------
    def get_summary_stats(self):
        """High-level survival and demographic statistics."""
        survived     = sum(1 for p in self.passengers if p.survived == 1)
        not_survived = sum(1 for p in self.passengers if p.survived == 0)
        ages         = [p.age  for p in self.passengers if p.age  is not None]
        fares        = [p.fare for p in self.passengers if p.fare is not None]

        # Survival by feature (from feature engineering analysis)
        survived_df     = self.df[self.df['Survived'] == 1]
        not_survived_df = self.df[self.df['Survived'] == 0]

        return {
            'total_passengers':       len(self.passengers),
            'survived':               survived,
            'did_not_survive':        not_survived,
            'survival_rate_%':        round(survived / len(self.passengers) * 100, 2),
            'average_age':            round(sum(ages)  / len(ages),  2) if ages  else None,
            'average_fare':           round(sum(fares) / len(fares), 2) if fares else None,
            'avg_family_size_survived':     round(survived_df['FamilySize'].mean(), 2),
            'avg_family_size_not_survived': round(not_survived_df['FamilySize'].mean(), 2),
            'alone_rate_survived_%':        round(survived_df['IsAlone'].mean() * 100, 2),
            'alone_rate_not_survived_%':    round(not_survived_df['IsAlone'].mean() * 100, 2),
        }

    # ----------------------------------------------------------
    # STEP 6: Export to JSON
    # ----------------------------------------------------------
    def to_json(self, filename=JSON_FILE):
        """Export everything — metadata, stats, missing values, passengers — to JSON."""
        filename = Path(filename)
        filename.parent.mkdir(parents=True, exist_ok=True)  # ensure folder exists

        data = {
            'metadata': {
                'dataset_name':     'Titanic Passenger Dataset',
                'export_date':      datetime.now().isoformat(),
                'total_passengers': len(self.passengers),
                'survival_rate_%':  round(self.df['Survived'].mean() * 100, 2),
                'source':           'Kaggle Titanic Dataset',
            },
            'summary_statistics':  self.get_summary_stats(),
            'descriptive_statistics': self.get_statistics(),
            'missing_values':      self.get_missing_values(),
            'passengers':          [p.to_dict() for p in self.passengers],
        }

        with open(filename, 'w') as f:
            json.dump(data, f, indent=2)

        print(f"✅ JSON exported to: {filename.resolve()}")
        return data


# ============================================================
# RUN
# ============================================================
if 'df' in locals() and not df.empty:
    dataset = TitanicDataset(df)

    print("\n" + "="*50)
    print("       TITANIC DATASET SUMMARY")
    print("="*50)
    for key, value in dataset.get_summary_stats().items():
        print(f"  {key.replace('_', ' ').title():<40} {value}")
    print("="*50)

    # Export to JSON
    dataset.to_json(JSON_FILE)
```
